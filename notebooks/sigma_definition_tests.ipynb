```python
import torch
import matplotlib.pyplot as plt
```

```python
c = torch.arange(0, 1, 0.01)
delta_t = 0.25
standard_dist = torch.distributions.normal.Normal(torch.zeros_like(c), torch.ones_like(c))
sigma = torch.sqrt(delta_t / (2 * standard_dist.icdf((1 + c) / 2)))
```

```python
plt.figure(figsize=(8, 6))
plt.plot(c, sigma, color='blue', linestyle='-', linewidth=2)
plt.plot(c, delta_t/2-torch.log(c), color='red', linestyle='-', linewidth=2)
plt.plot(c, -torch.log(c), color='yellow', linestyle='-', linewidth=2)
plt.axhline(delta_t, color='red')
# Add labels and title
plt.title("title")
plt.xlabel("confidence")
plt.ylabel("sigma")

# Add a grid for better readability
plt.grid(True, linestyle='--', alpha=0.6)

# Display the plot
plt.show()
```
